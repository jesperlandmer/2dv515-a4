# Scatterplot Presentation of the Dataset

```python
import csv

X = []
Y = []

with open('spiral.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count != 0:
            X.append([float(row[0]), float(row[1])])
            Y.append(int(row[2]))
        line_count+=1
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

# Define colors
def zero():
    return "red"

def one():
    return "green"

def two():
    return "blue"

# Color switch
colors = { 0:zero, 1:one, 2:two }


# Create scatter
fig, ax = plt.subplots()
classes = [0, 1, 2]
for c in classes:
    x = []
    y = []
    color = colors.get(c)
    for i in range(len(X)):
        if Y[i] == c:
            x.append(X[i][0])
            y.append(X[i][1])
    ax.scatter(x, y, c=color(), label=c, alpha=0.6, edgecolors='none')

# Plot scatter with legend
ax.legend()
plt.show()
```
